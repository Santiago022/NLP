```python
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv(r"C:\Users\SLR02\Documents\Trabajos\GIE\DATOS_PAPER_HUE The hourly usage of energy dataset forbuildings in British Columbia\Residencia 22/Residential22_procesing_15min.csv",sep=',')
data.Horodate=pd.to_datetime(data.Horodate)
data=data.set_index('Horodate')
data.head()
```

```python
data
```

```python
NewData2 = pd.DataFrame(index=data.index, columns=['Energia'])
NewData2.Energia=data.Potencia/4
```

```python
NewData2['Hora']=NewData2.index.hour
NewData2['Mes']=NewData2.index.month
NewData2['DiaSemana']=NewData2.index.dayofweek+1
NewData2['Laboralidad']=1
NewData2['Potencia']=data.Potencia
NewData2['L']=(NewData2.DiaSemana==1).astype(int)
NewData2['M']=(NewData2.DiaSemana==2).astype(int)
NewData2['Mi']=(NewData2.DiaSemana==3).astype(int)
NewData2['J']=(NewData2.DiaSemana==4).astype(int)
NewData2['V']=(NewData2.DiaSemana==5).astype(int)
NewData2['S']=(NewData2.DiaSemana==6).astype(int)
NewData2['D']=(NewData2.DiaSemana==7).astype(int)
##MESES
NewData2['Ene']=(NewData2.Mes==1).astype(int)
NewData2['Feb']=(NewData2.Mes==2).astype(int)
NewData2['Mar']=(NewData2.Mes==3).astype(int)
NewData2['Abr']=(NewData2.Mes==4).astype(int)
NewData2['Mayo']=(NewData2.Mes==5).astype(int)
NewData2['Jun']=(NewData2.Mes==6).astype(int)
NewData2['Jul']=(NewData2.Mes==7).astype(int)
NewData2['Ago']=(NewData2.Mes==8).astype(int)
NewData2['Sep']=(NewData2.Mes==9).astype(int)
NewData2['Oct']=(NewData2.Mes==10).astype(int)
NewData2['Nov']=(NewData2.Mes==11).astype(int)
NewData2['Dic']=(NewData2.Mes==12).astype(int)
###
NewData2['LAB_L']=(NewData2.Laboralidad==1).astype(int)
NewData2['LAB_SL']=(NewData2.Laboralidad==2).astype(int)
NewData2['LAB_NL']=(NewData2.Laboralidad==3).astype(int)
```
