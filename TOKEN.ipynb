## LECTURA DE TODAS LAS ETIQUETAS DEL CORPUS 

```python
b= []
aux_c=[]
pala_base=[]
b.append("<s>")
aux_c.append(" ")
aux=0
for i in range (2,len(a)):
    if aux > (len(a)):
        break
    if len(a[aux].split())>4 or len(a[aux].split())==1:
        aux=aux+1
    elif len(a[aux].split())==4:
        b.append(a[aux].split()[2])
        pala_base.append(a[aux].split()[0])
        aux_c.append(a[aux].split()[0])
    elif len(a[aux].split())==0:
        b.append("<\s>")
        b.append("<s>")
        aux_c.append(" ")
        aux_c.append(" ")
    aux=aux+1
print(b)
```

## OBTENER SOLO LOS CARACTERES UNICOS DE LA LISTA ANTERIOR

```python
unico=set(b)
```

```python
print(len(a),len(b),len(unico),len(aux_c))
```

## CREAR UNA LISTA DE ESTOS CARACTERES UNICOS 

```python
unicolist = list(unico)
```

```python
unicolist
```

## CREACCION DE UNA TABLA DE FRECUENCIA PARA  CALCULAR LA PROBABILIDAD DE TRANSICION, QUE SEA EL NUMERO DE VECES QUE SE REPITE UNA ETIQUETA SEGUIDA DE OTRA

```python
frecuencias=[[0]*len(unicolist)]*len(unicolist) #UNA MATRIZ CUADRADA DE TAMANO DE LOS CARACTERES UNICOS
for ki in range (len(unicolist)): # UN BARRIDO PARA LAS FILAS
    aux2=[0]*len(unicolist)     # UNA LISTA QUE ALMACENE LAS REPETICIONES PARA CADA POSICION DE LAS FILAS
    for j in range(len(unicolist)):  # UN BARRIDO DE COLUMNAS
        aux=0                        # UN CONTADOR QUE ALMACENE LA CANTIDAD DE REPETICIONES DE UNA PALABRA DESPUES DE LA ANALIZADA
        for i in range (len(b)-1):   # UN BARRIDO DE TODO EL CORPUS DE ENTRENAMIENTO
            if b[i] == unicolist[ki] and b[i+1] == unicolist[j]: # COMPARACION DE TODOS LOS ELEMENTOS UNICOS CONTRA CADA POSICION DE LA FILA
                aux+=1                                          # SI ES VERDADERO ENTONCES AUMENTA CONTADOR
        aux2[j]=aux                                            # acabado el barrido, colocar en cada posicion de la columna
    frecuencias[ki]=aux2                                        # acabado el barrido igular a cada posicion de la fila
print(frecuencias[0])                                           # imprimir la primera posicion de la lista
```

## COMPARO SI ESTA CORRECTO LOS ITERADORES DE LOS BARRIDOS 

```python
[i,j,ki,len(b)]
```

## CREAR UNA LISTA QUE ALMACENE LAS CANTIDADES DE REPETICIONES DE CARACTERES UNICOS 

```python
frecuenciaPalab = []
for i in range(len(unicolist)):
    frecuenciaPalab.append(b.count(unicolist[i]))
```

## COMPROBAR TAMANOS DE LISTAS DE FRECUENCIAS

```python
print(len(frecuenciaPalab),len(frecuencias))
```

## DEL TEOREMA DE BAYES, RELACIONANDO VEROSIMILITUD SERIA ANALOGA A DECIR LA MATRIZ DE PROBABILIDAD DE TRANSICION YO LE NOMBRARE VEROLIMILITUD PARA NO CONOFUNDIRME

```python
Verosimilitud=[[0]*len(unicolist)]*len(unicolist)
for i in range(len(frecuencias)):
    Verosimilitud[i] = [x / frecuenciaPalab[i] for x in frecuencias[i]]
```

```python
import pandas as pd 
df = pd.DataFrame(Verosimilitud,index =unicolist, columns = unicolist) 
df2 = pd.DataFrame(frecuencias,index =unicolist, columns = unicolist) 
```
